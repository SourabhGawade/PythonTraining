```python
import re
import pandas as pd
```

```python
pattern = re.compile(r'finished enrichment')
#print(re.compile(r'finished enrichment'))
file = "S:\\1683428\\Learning\\TCS Training\\Python Training\\smaple_info.txt"
mydataset ={
    'col1': [],
    'col2': [],
    'col3': [],
    'col4': []
}
myvar = pd.DataFrame(mydataset)
for line in open(file):
    lines = pattern.search(line)
    string_to_add = line[lines.span()[0]:].split('; ')
    myvar.loc[len(myvar.index)] = [string_to_add[0], string_to_add[1], string_to_add[2], string_to_add[3]]

print(myvar)
```
